### Combine Dataframe by .merge and .concat


2.6 Merge


```python
import pandas as pd

superstore_order = pd.read_csv("Superstore/superstore_order.csv")
superstore_people = pd.read_csv("Superstore/superstore_people.csv")
superstore_return = pd.read_csv("Superstore/superstore_return.csv")
```

```python
superstore_order.merge(
    superstore_return[superstore_return["Returned"] == "Yes"], on="Order ID", how="inner"
)[["Customer ID", "Returned"]].drop_duplicates()
```

**Q4 – What does `how="inner"` do?**

Ans: It keeps only rows where the same Order ID appears in both tables. So we only see orders that were returned.


**Q5 – What does the result above tell you?**

Ans: It is the list of customers (Customer IDs) who had at least one order returned.


2.7) Concatenate


```python
pd.concat([superstore_order, superstore_people], axis=1, join="inner")
```

**Q6 – What is the difference between `inner` and `outer` in `pd.concat`'s `join`?**

Ans: Inner keeps only columns that exist in all DataFrames. Outer keeps all columns from every DataFrame and fills missing with NaN.


**Q7 – Append a new row to superstore_order. Set Row ID to your student ID and fill other fields like existing rows.**


```python
new_row = superstore_order.iloc[0].copy()
new_row["Row ID"] = 99999
new_row["Order ID"] = "XX-9999-99999"
superstore_order = pd.concat([superstore_order, new_row.to_frame().T], ignore_index=True)
superstore_order.tail(2)
```

### Groupby


```python
superstore_order["Profit Ratio"] = superstore_order["Profit"] / superstore_order["Sales"]

superstore_order.groupby(["Category", "Sub-Category"]).agg(
    mean_profit_ratio=("Profit Ratio", "mean")
)
```

### Pivot and Melt


Pivot


```python
superstore_order.pivot_table(
    index="State", columns="Ship Mode", values="Order ID", aggfunc="count"
).fillna(0).head(10)
```

```python
pivot_table_result = superstore_order.pivot_table(
    index="State", columns="Ship Mode", values="Order ID", aggfunc="count"
).fillna(0)
print(pivot_table_result)
```

Melt


```python
melted_result = pd.melt(
    pivot_table_result.reset_index(),
    id_vars=["State"],
    var_name="Ship Mode",
    value_name="Order Count",
)
print(melted_result)
```

**Q9 – What is the advantage of using `melt`?**

Ans: It turns a wide table (many columns) into a long one (fewer columns, more rows), so it is easier to plot or filter by that variable.

**Q12 – Complete the function so the `ratio` column is First Class / Standard Class on `pivot_table_result`.**


```python
# function to transform the ratio
def get_class_ratio(row):
    first_class = row["First Class"]
    standard_class = row["Standard Class"]
    ratio = first_class / standard_class if standard_class != 0 else 0
    return ratio


pivot_table_result["ratio"] = pivot_table_result.apply(get_class_ratio, axis=1)

pivot_table_result.head()
```

**Q13 – What does the `apply` function do?**

Ans: It runs our function on every row (or column) and returns a new Series with the results. Here it computes the ratio for each row.


**Q14 – Create a column `short_ratio` that does the same as Q12 but with a lambda.**


```python
pivot_table_result["short_ratio"] = pivot_table_result.apply(
    lambda row: row["First Class"] / row["Standard Class"] if row["Standard Class"] != 0 else 0,
    axis=1,
)

pivot_table_result.head()
```

[Q15] What is the difference between `apply` and `lambda` function? give 2 examples use case.

**Answer:** `apply` is a **method** that applies a function along an axis (rows or columns); `lambda` is a way to define a **small anonymous function**. They are complementary: you often pass a `lambda` (or a named function) to `apply`.

**Two use cases:**

1. **Row-wise calculation with lambda:** e.g. `df.apply(lambda row: row["A"] / row["B"], axis=1)` for a quick per-row ratio.
2. **Reusable logic with a named function:** e.g. `df.apply(get_class_ratio, axis=1)` when the logic is more complex or reused (as in Q12).
