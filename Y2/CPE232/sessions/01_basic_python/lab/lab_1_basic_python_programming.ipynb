## [5] OOP


```python
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        """Initialize attributes to describe a car."""
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def get_description_name(self):
        """Return a neatly formatted descriptive name."""
        long_name = f"{self.year} {self.make} {self.model}"
        return long_name.title()

    def read_odometer(self):
        """Print a statement showing the car's mileage."""
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        """Set the odometer reading to the given value."""
        self.odometer_reading = mileage
```

```python
my_car = Car("Honda", "Civic", 2016)
my_car.read_odometer()

# update odometer
my_car.update_odometer(50)
my_car.read_odometer()
```

[Q2] From the above cell, explain in your own words how the output came about?

Ans: The expression `arr_a[(arr_a > 5) & (arr_a < 10)]` uses boolean indexing. First, `(arr_a > 5)` creates a boolean array where True indicates elements greater than 5. Similarly, `(arr_a < 10)` creates a boolean array where True indicates elements less than 10. The `&` operator performs element-wise AND operation on these two boolean arrays, resulting in True only where both conditions are satisfied (elements between 5 and 10, exclusive). Finally, this boolean array is used to index `arr_a`, returning only the elements that satisfy both conditions: [6, 7, 8, 9].

```python
arr_a[(arr_a > 5) and (arr_a < 10)]
```

[Q3] Explain in your own words why the above cell gives an error.

Ans: The error occurs because Python's `and` operator does not work with NumPy arrays. The `and` operator expects boolean values (True/False), but `(arr_a > 5)` returns a NumPy boolean array, not a single boolean value. When Python tries to evaluate `(arr_a > 5) and (arr_a < 10)`, it attempts to use the truth value of the entire array, which raises a ValueError because arrays with more than one element cannot be evaluated as a single boolean value. The `and` operator is designed for scalar boolean operations, not element-wise array operations.


[Q4] And what should be written instead so that the code is error-free?

Ans: Instead of using `and`, we should use the bitwise `&` operator: `arr_a[(arr_a > 5) & (arr_a < 10)]`. The `&` operator performs element-wise AND operation on NumPy boolean arrays, which is the correct way to combine multiple boolean conditions for array indexing. Additionally, each condition should be wrapped in parentheses to ensure proper operator precedence: `(arr_a > 5) & (arr_a < 10)`.

[Q5] Summarize the value of the argument _axis_, what is the value for row-wise summation and column-wise summation, respectively?

Ans: The `axis` argument specifies along which dimension the aggregation operation is performed. For a 2D array:

- `axis=0` performs operations along the first dimension (down the columns), resulting in column-wise aggregation. This collapses rows and keeps columns, so for a 3Ã—4 array, it produces 4 values (one per column).
- `axis=1` performs operations along the second dimension (across the rows), resulting in row-wise aggregation. This collapses columns and keeps rows, so for a 3Ã—4 array, it produces 3 values (one per row).
  In summary: `axis=0` for column-wise, `axis=1` for row-wise.

[Q6] What is the difference between .iloc and .loc?

Ans: The main differences are:

- `.iloc` uses integer-based indexing (position-based), similar to list indexing. It accesses data by the integer position of rows and columns (0-based indexing). For example, `df.iloc[0:5, 1:3]` selects rows 0-4 and columns 1-2 by their positions.
- `.loc` uses label-based indexing (name-based). It accesses data by the actual index labels and column names. For example, `df.loc[0:5, 'age':'sex']` selects rows with index labels 0-5 and columns from 'age' to 'sex' (inclusive).
  Additionally, `.loc` is inclusive on both ends when using slice notation (e.g., `df.loc[0:5]` includes both 0 and 5), while `.iloc` follows Python's standard slicing (exclusive on the right end, e.g., `df.iloc[0:5]` includes 0-4 but not 5).

[Q7] In the above cell, Explain why pandas .agg()/.apply() method accepts 'len' and 'lambda' without quotation marks.

Ans: Pandas `.agg()` and `.apply()` methods can accept both string names of functions (like `'len'` or `'mean'`) and actual function objects (like `len` or `lambda` functions) without quotation marks. When you pass `len` without quotes, you're passing the function object itself, which pandas will call on each element or group. Similarly, `lambda` functions are function objects that can be passed directly. Pandas is flexible and accepts either format: string names (e.g., `'len'`, `'mean'`) for built-in aggregation functions, or function objects (e.g., `len`, `lambda x: x.mode()[0]`) for custom operations. This design allows for both convenience (using string shortcuts) and flexibility (using custom functions).
